# Introduction

Hello, welcome to BaseAL 👋

BaseAL is a **modular framework for evaluating active learning methods** allowing you to:
- Setup multiple sampling experiments in parallel
- Easily add and adapt methods and parameters
- Visualise the embedding space and uncertainty during training

While the active learning framework itself is modality agnostic (only requiring labeled embeddings), embedding generation **currently only supports audio data** generated using [bacpipe](https://github.com/bioacoustic-ai/bacpipe), developed by [Vincent Kather](https://github.com/vskode).

*BaseAL is not an annotation tool*, it is designed to evaluate sampling methods on prelabeled datasets. You are welcome to extend this tool for annotation and adapt it for your own data pipelines 🌱

If you have any questions, don't hesistate to get in [contact](mailto:benmcewen@outlook.com).
