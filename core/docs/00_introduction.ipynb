# Introduction

BaseAL is a **modular framework for setting up and evaluating active learning methods** allowing you to:
- Setup multiple sampling experiments in parallel
- Easily add and adapt methods and parameters
- Visualise the embedding space during training

BaseAL is built on top of [bacpipe](https://github.com/bioacoustic-ai/bacpipe), developed by [Vincent Kather](https://github.com/vskode). While the active learning framework itself is modality independent (only requiring labeled embeddings), embedding generation **currently only supports audio data** üê§

*BaseAL is not an annotation tool*, it is designed to evaluate sampling methods on prelabeled datasets. You are welcome to extend this tool for annotation and adapt it for your own data pipelines üå±
